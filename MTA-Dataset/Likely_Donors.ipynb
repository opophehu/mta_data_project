```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

blocks = pd.read_csv('census_block_loc.csv')
census = pd.read_csv('nyc_census_tracts.csv',index_col=0)
```

```python
census.head(20)
```

```python
#Income per capita by borough

incomePerCap = census.groupby(['Borough'])['IncomePerCap'].mean().reset_index()
plt.bar(incomePerCap['Borough'], incomePerCap['IncomePerCap'], width = 0.6, color='k')
```

```python
#Female %

gender = census.groupby(['Borough'])['Men', 'Women'].sum().reset_index()

gender['PerWomen'] = gender['Women']/(gender['Women'].sum()*100)
```

```python
plt.bar(gender['Borough'],gender['PerWomen'],width=0.6, color = 'k')
```
